#### Importando bibliotecas

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

#### Se cargan los set de datos en las variables df1 y df2

```python
df1 = pd.read_csv('data/w_data.csv', sep=';')
df1
```

```python
df2 = pd.read_csv('data/p_data.csv', sep=';')
df2
```

#### se hace un merge de los datos se considera usar un left ya que el primer dataset tiene mas filas que el segundo 

```python
df_merge = pd.merge(df1,df2, on='ID', how='left', indicator=True)
df_merge
```

#### 2)Corregir los target a solo dos valores distintos ‘<=50k’ y ‘>50k’

#### revisar los diferentes valores de target

```python
df_merge.Target.value_counts()
```

#### se usa un replace para cambiar los valores

```python
df_merge['Target'] = df_merge['Target'].replace('<=50K.','<=50K',regex=True)
df_merge['Target'] = df_merge['Target'].replace('>50K.','>50K',regex=True)
df_merge.Target.value_counts()
```

#### 3)Calcular la proporción que hay entre personas con salarios >50k respecto del total para personas de distinto sexo, Raza y Educacion. Por ejemplo cuál es el porcentaje de mujeres con un salario mayor a 50K y compararlo con el de los hombres. Lo mismo para personas de distintas razas y Educación

#### Se hace la primera agrupacion usando las columnas 'Target' y 'Sexo', se usa count() para obtener la sumatoria de filas por cada segmento

```python
agrupacionPorGenero = df_merge.groupby([df_merge['Target'],df_merge['Sex']]).count()
agrupacionPorGenero
```

#### se puede usar cualquiera de las otras columnas para realizar un calculo matematico del porcentaje 

```python
grupoGeneroPorcentaje = agrupacionPorGenero['Workclass']/agrupacionPorGenero['Workclass'].sum() *100
grupoGeneroPorcentaje
```

#### Mujeres con un salario mayor a 50k: 3.62%, hombres con un salario mayor a 50k: 20.30%

#### Se hace la segunda agrupacion usando las columnas 'Target' y 'Race', se usa count() para obtener la sumatoria de filas por cada segmento

```python
agrupacionPorRaza = df_merge.groupby([df_merge['Target'],df_merge['Race']]).count()
agrupacionPorRaza 
```

#### se puede usar cualquiera de las otras columnas para realizar un calculo matematico del porcentaje 

```python
grupoRazaPorcentaje = agrupacionPorRaza['Workclass']/agrupacionPorRaza['Workclass'].sum() *100
grupoRazaPorcentaje
```

#### Amer-Indian-Eskimo con sueldo mayor a 50k: 0.11%, Asian-Pac-Islander con sueldo mayor a 50k: 0.83%, Black con sueldo mayor a 50k: 8.43%, Otros con sueldo mayor a 50k: 0.72%, Blancos con sueldo mayor a 50k: 21.71%

#### Se hace la tercera agrupacion usando las columnas 'Target' y 'Age', se usa count() para obtener la sumatoria de filas por cada segmento

```python
agrupacionPorEdad = df_merge.groupby([df_merge['Target'],df_merge['Age']]).count()
agrupacionPorEdad 
```

#### se puede usar cualquiera de las otras columnas para realizar un calculo matematico del porcentaje 

```python
grupoEdadPorcentaje = agrupacionPorEdad['Workclass']/agrupacionPorEdad['Workclass'].sum() *100
grupoEdadPorcentaje
```
